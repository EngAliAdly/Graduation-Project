```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
sw = open("stop_words_english.txt",'r',encoding='utf-8').read().split('\n')
sw
```

```python
DescriptionDF['value'].values.tolist()[0]
```

```python
def remove_sw(text):
    words = text.lower().split()
    clean_words = []
    for word in words:
        if word not in sw:
            clean_words.append(word)
        
        
    return ' '.join(clean_words)
```

```python
import re
def denoise(text):
    return re.sub('[^a-zA-Z ]*','',text).strip()
```

```python
ps = PorterStemmer()
```

```python
def stemming(text):
    words = text.lower().split()
    stemmed_text=[]
    for word in words:
        stemmed_text.append(ps.stem(word))
    return stemmed_text
```

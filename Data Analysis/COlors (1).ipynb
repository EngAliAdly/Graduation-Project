```python
df1 = df.dropna(subset=['colours'])
df1.info()
```

```python
df1
```

```python
TBDF_ProductAttributes=[ast.literal_eval(df1.iloc[[0]].colours.values.tolist()[0])]
DF_ProductAttributes=pd.DataFrame(TBDF_ProductAttributes)
DF_ProductAttributes
```

```python
colorsDF=pd.DataFrame(DF_ProductAttributes.iloc[[0]].colors.values[0]).T
colorsDF=colorsDF.reset_index(drop=True)
colorsDF
```

```python
parent_id=[]
for i in range(len(colorsDF)):
    parent_id.append(df1.iloc[[0]].id.values[0])
parent_id={'parent_id':parent_id}
parent_id_DF=pd.DataFrame(parent_id)
colorsDF=pd.concat([colorsDF,parent_id_DF],axis=1)
```

```python
colorsDF
```

```python
AllcolorsDF=colorsDF
AllcolorsDF
```

```python
for i in range(1,len(df1)):
    TBDF_ProductAttributes=[ast.literal_eval(df1.iloc[[i]].colours.values.tolist()[0])]
    DF_ProductAttributes=pd.DataFrame(TBDF_ProductAttributes)
    colorsDF=pd.DataFrame(DF_ProductAttributes.iloc[[0]].colors.values[0]).T
    colorsDF=colorsDF.reset_index(drop=True)
    parent_id=[]
    for j in range(len(colorsDF)):
        parent_id.append(df1.iloc[[i]].id.values[0])
    parent_id={'parent_id':list(map(int, parent_id))}
    parent_id_DF=pd.DataFrame(parent_id)
    colorsDF=pd.concat([colorsDF,parent_id_DF],axis=1)
    AllcolorsDF=AllcolorsDF.append(colorsDF)
AllcolorsDF
```

```python
AllcolorsDF=AllcolorsDF.reset_index(drop=True)
AllcolorsDF.info()
```

```python
AllcolorsDF=AllcolorsDF.astype({'parent_id': 'int32'})
```
